```python
import numpy as np
import pandas as pd

#Extraer la información del archivo.
Documento = pd.read_csv("/content/country_vaccinations.csv")
```

```python
#Mostrar la estructura y tipos de datos de cada columna.
Documento['date'] = pd.to_datetime(Documento['date'])
Documento.info()
```

```python
#Determinar la cantidad de vacunas aplicadas de cada compañía.
ConteoVacunas = Documento[['daily_vaccinations','vaccines']].groupby('vaccines').sum()
ConteoVacunas
```

```python
#Obtener la cantidad de vacunas aplicadas en todo el mundo.
print(Documento['daily_vaccinations'].sum())
```

```python
#Calcular el promedio de vacunas aplicadas por país.
print(Documento[['daily_vaccinations','country']].groupby('country').mean())
```

```python
# Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
mask = Documento['date'] == '29/01/21'
filtrado = Documento.loc[mask]
print(filtrado['daily_vaccinations'].sum())
```

```python
#Crear un dataframe nuevo denominado conDiferencias que contenga los datos
#originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.
conDiferencias = Documento
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

```python
#Periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.
print("La fecha mas antigua es:", Documento['date'].min())
print("La fecha mas reciente es:", Documento['date'].max())
print("Periodo de tiempo transcurrido entre ambas fechas:", (Documento['date'].max() - Documento['date'].min()))
```

```python
#Crear un dataframe nuevo denominado conCantidad que contenga los datos originales
#y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).
conCantidad = Documento
conCantidad['canVac'] = conCantidad['daily_vaccinations']
#No comprendo lo que se pide, la cantidad de vacunas utilizadas cada dia ya vienen en la columna daily_vaccinations
```

```python
#Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.
mask = Documento['date'] < '20/12/20'
antes20 = Documento[mask]
antes20
```

```python
#Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.
serie = Documento['vaccines']
arr = [True, False]
mask = np.random.choice(arr, size=serie.count())
for i in range(serie.count()):
  mask[i] = 'Pfizer/BioNTech' in serie[i]
pfizer = Documento[mask]
pfizer
```

```python
#Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer)
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```

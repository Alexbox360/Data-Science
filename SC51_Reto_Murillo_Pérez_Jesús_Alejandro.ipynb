```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statistics
from mpl_toolkits.mplot3d import axes3d
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
Dt_meses = pd.read_csv("/content/0.3z6dlwbp9zv0.iz9i49q40yhreto_1.csv", encoding='latin-1')
Dt_entidades = pd.read_csv("/content/0.zb64r0glsn90.rxrfxpdwg2breto_3.csv", encoding='latin-1')
```

```python
fig = sns.boxplot(data=Dt_entidades, y="Autobus", color = "Green")
plt.xlabel("Acumulativo de las entidades estudiadas")
plt.ylabel("Conteo de autobuses")
plt.axhline(statistics.median(Dt_entidades["Autobus"]), dashes=(2, 2))
plt.annotate("Mediana", (0.1,statistics.median(Dt_entidades["Autobus"])))
print("mediana = ", statistics.median(Dt_entidades["Autobus"]))
plt.title("boxplot del conteo de autobuses")
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Dt_entidades["Automovil"], Dt_entidades["Autobus"], Dt_entidades["Camioneta"], marker='o', color="red")
ax.set_xlabel("Automovil")
ax.set_ylabel("Autobus")
ax.set_zlabel("Camioneta")
plt.title("Grafico 3D")
```

```python
X = Dt_entidades[["Autobus","Camioneta"]]
y = Dt_entidades["Automovil"]
reg = LinearRegression().fit(X, y)
y_pred = reg.predict(X)
sns.residplot(x=y, y=y_pred, color="red")
plt.ylabel("Error del ajuste")
plt.title("Modelo residual")
plt.show()
```

```python
y = Dt_entidades["Autobus"]
X = Dt_entidades["Automovil"]
fig, ax = plt.subplots(2,figsize=(8,8))
plt.subplots_adjust(hspace=0.45)
sns.regplot(ax=ax[0], x=X,y=y)
sns.residplot(ax=ax[1], x=X, y=y, color="red")
ax[0].set_title("Regresion lineal")
ax[1].set_title("Modelo residual")
fig.suptitle('Dependecia de la variable autobus con respecto a los automoviles')
plt.show()
#Pese a que los valores para los finales para la regresion son consistentes con lo predecido la relacion en un inicio es poco visible debido a la cercania de los puntos
X = Dt_entidades["Autobus"]
y = Dt_entidades["Automovil"]
fig1, ax1 = plt.subplots(2,figsize=(8,8))
plt.subplots_adjust(hspace=0.45)
sns.regplot(ax=ax1[0], x=X,y=y)
sns.residplot(ax=ax1[1], x=X, y=y, color="red")
ax1[0].set_title("Regresion lineal")
ax1[1].set_title("Modelo residual")
fig1.suptitle('Dependecia de la variable autobus con respecto a los automoviles')
#Pese a lo anterior, si tratamos a los automoviles como variable dependiente, en esta ocasion la relacion lineal que guardan
#estas varibles es mucho ms clara, a partir de este analisis, podemos asumir que si existe realcion entre las varibales
#a mayor numero de autobuses, el nuemro de automoviles se ve ligeramente reducido
```

```python
y = Dt_entidades["Autobus"]
X = Dt_entidades["Automovil"]
y = y[:6]
X = X[:6]
sns.kdeplot(x=X,y=y,fill="True",color="Green")
plt.title("Curva de densidad de probabilidad Autobus vs Automovil")
plt.show()
fig, ax = plt.subplots()
sns.kdeplot(ax=ax, x=X, fill="True", color="Red", alpha=0.5)
sns.kdeplot(ax=ax, x=y, fill="True", color="Green", alpha=0.5)
ax.legend(labels=["Autobus", "Automovil"], title="Vehiculo",  loc=(.88,1.05))
plt.title("Curva de densidad de probabilidades")
```

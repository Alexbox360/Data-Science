```python
# se importan las libreras
from pylab import *
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#Domio en el cual se trabajara, (valores los culaes 'x' y 'y' son capaces de tomar)
xmax = 2
xmin = -xmax
ymax = 1
ymin = -ymax
# Nmero de intervalos en los cuales se de dividiran los ejes X y Y
EjeX = 20
EjeY = 20
# Se utiliza la funcin linspace para calcular el espaciamiento de los puntos en la grafica, dados un intervalo [xmin,xmax] seguido por el nmero de puntos los cuales se deben de encontrar en este intervalo EjeX
x = linspace(xmin,xmax,EjeX)
y = linspace(ymin,ymax,EjeY)
#La funcin meshgrid permite crear una matriz de las coordenadas 'x' y 'y' dadas anteriormente, obteniendo as una malla para luego poder evaluar una funcion de dos variables
X, Y = meshgrid(x, y)
#Se escriben las componenetes para las coordenadas i y j de E(x,y) las cuales harn uso de las matrices obtenidas por la funcion meshgrid para su evaluacion punto a punto en el espacio coordenado (x,y)
a = 0.3
d = 1
q = 2
k = (4 * math.pi *  8.8541878176*(10**-12))
Ex = (q / k) * ( ( (X - 1) / (X*X + Y*Y + d*d - 2*X)**(3 / 2) ) - ( (a*d*(d*X - a*a) ) / (d*d*X*X + d*d*Y*Y - 2*a*a*d*X + a**4)**(3 / 2) ) )
Ey = (q / k) * ( (Y / (X*X + Y*Y + d*d - 2*X )**(3 / 2) ) - ( (a*d*d*Y) / (d*d*Y*Y + d*d*X*X + a**4 - 2*a*a*d*X )**(3 / 2) ) )
E = sqrt(Ex*Ex + Ey*Ey)
U = Ex / E
V = Ey / E
#Se crea una figura en la cual se graficara la funcin
figura, h = plt.subplots(figsize=(12, 12))
#Dibujaremos el campo vectoria usando la funcin quiver()
color = E
Q = quiver(X, Y, U, V, color)
quiverkey(Q, 0.85, 1.05, 1, '1mT',labelpos='N')
dx = (xmax - xmin)/(EjeX-1)
dy = (ymax - ymin)/(EjeY-1)
axis([xmin-dx, xmax+dx, ymin-dy, ymax+dy])
h.plot(a*a / d, 0, marker="o", color="blue")
Drawing_colored_circle = plt.Circle(( d , 0 ), 0.08, color = 'red' )
h.set_aspect( 1 )
h.add_artist( Drawing_colored_circle )
Drawing_colored_circle = plt.Circle(( 0 , 0 ), a, fill = False )
h.set_aspect( 1 )
h.add_artist( Drawing_colored_circle )
#Colocando nombres a los ejes
title('Campo vectorial')
xlabel('x')
ylabel('y')
show()
```
